```python
import matplotlib.pyplot as plt
names=['Ajay','Ankur','Amit','Raj','Rahul']
marks=[350,450,432,458,582]
plt.plot(names,marks,linestyle='--',linewidth=4,color='red')
plt.xlabel('Name')
plt.ylabel('marks')
plt.title('marksheet')
plt.show()
```

```python
import matplotlib.pyplot as plt
names=['Ajay','Ankur','Amit','Raj','Rahul']
marks=[350,450,432,458,582]
plt.plot(names,marks,linestyle='--',linewidth=4,color='red',marker='o',markerfacecolor='green',markersize=8)
plt.xlabel('Name')
plt.ylabel('marks')
plt.title('marksheet')
plt.show()
```

```python
import matplotlib.pyplot as plt
name=[]
sal=[]
for i in range(1,6):
    a=input('enter name = ')
    b=int(input('enter salary = '))
    name.append(a)
    sal.append(b)
plt.plot(name,sal)
plt.xlabel('Name')
plt.ylabel('Salary')
plt.show()
```

```python
import matplotlib.pyplot as plt
s1=[1,3,5,5]
m1=[58,59,62,47]
s2=[2,4,6,8]
m2=[61,62,47,53]
plt.plot(s1,m1,color='blue',label='odd')
plt.plot(s2,m2,color='red',label='even')
plt.xlabel('sem')
plt.ylabel('marks')
plt.legend()
plt.show()
```
